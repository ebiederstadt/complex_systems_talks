```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def model(Y, t, r):
    x = Y[0]
    y = Y[1]

    return [
        -x + r * x * y - x ** 2,
        y * (1 - y)
    ]
```

```python
r = 2.5
t = np.linspace(0, 10, 100)
y0 = [0.5, 0.8]

solution = integrate.odeint(model, y0, t, args=(r,))
```

```python
plt.plot(solution[:, 0], label=r"$\dot{x}$")
plt.plot(solution[:, 1], label=r"$\dot{y}$")
plt.xlabel("t")
plt.legend()
plt.title(f"Commensalism Model (r={r})")
plt.show()
```

```python
x_values, y_values = np.meshgrid(np.arange(0, 50, 0.1), np.arange(0, 20, 0.1))
x_dot = -x_values + r * x_values * y_values - x_values ** 2
y_dot = y_values * (1 - y_values)

plt.figure(figsize=(10, 8))
plt.streamplot(x_values, y_values, x_dot, y_dot)
plt.xlabel("dx/dt")
plt.ylabel("dy/dt")
plt.scatter([r-1], [1], color="red")
plt.title("Phase Space with Stable Point shown in red")
plt.show()
```

```python
def f_1(r):
    return np.sqrt(r)

def f_2(r):
    return -np.sqrt(r)

domain = np.linspace(0, 10, 500)
plt.plot(domain, f_1(domain), "b-")
plt.plot(domain, f_2(domain), "r--")
plt.xlabel("r")
plt.ylabel(r"$x_{eq}$")
plt.grid()
plt.show()
```
