```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
# omega_dot = -g/L * sin(theta)
# theta_dot = omega
def model(y, t, L, g):
    theta, omega = y
    return [omega, -(g/L)*np.sin(theta)]

def gen_title(L, y0):
    return fr"Numerically Simulated Pendulum. L={L}, $\theta_0={y0[0]:.1f}$, $\omega_0={y0[1]:.1f}$"
```

```python
# Initial conditions
y0 = [np.pi - 0.1, 0.0]
L = 0.5

t = np.linspace(0, 10, 1000)

solution = integrate.odeint(model, y0, t, args=(L,9.81))
print(solution.shape)
```

```python
plt.plot(t, solution[:, 0], label=r"$\theta(t)$")
plt.plot(t, solution[:, 1], label=r"$\omega(t)$")
plt.legend()
plt.xlabel("t")
plt.title(gen_title(L, y0))
```

```python
y0 = [np.pi - 0.1, 0.4]
L = 0.5

t = np.linspace(0, 10, 1000)

solution = integrate.odeint(model, y0, t, args=(L,9.81))
plt.plot(t, solution[:, 0], label=r"$\theta(t)$")
plt.plot(t, solution[:, 1], label=r"$\omega(t)$")
plt.legend()
plt.xlabel("t")
plt.title(gen_title(L, y0))
```

```python
# Phase space
xvalues, yvalues = np.meshgrid(np.arange(-8, 8, 0.1), np.arange(-3, 3, 0.1))
g, L = 1, 1
x_dot = yvalues
y_dot = -(g/L) * np.sin(xvalues)


plt.figure(figsize=(9, 6))
plt.streamplot(xvalues, yvalues, x_dot, y_dot)
plt.title("Phase Space for the Pendulum (g=1.0, L=1.0)")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$\omega$")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
y0 = [0.0, 1.0]
t = np.linspace(0, 500, 2000)
ys = integrate.odeint(model, y0, t, args=(1.0,1.0))

y0 = [0.0, 2.0]
ys_2 = integrate.odeint(model, y0, t, args=(1.0,1.0))

y0 = [-2 * np.pi, 3.0]
ys_3 = integrate.odeint(model, y0, t, args=(1.0,1.0))

plt.figure(figsize=(9, 6))
plt.streamplot(xvalues, yvalues, x_dot, y_dot)
plt.plot(ys[:,0], ys[:, 1], label=r"$\omega=1.0$")
plt.plot(ys_2[:,0], ys_2[:, 1], label=r"$\omega=2.0$")
plt.plot(ys_3[:,0], ys_3[:, 1], label=r"$\omega=3.0$")
plt.title("Phase Space for the Pendulum (g=1.0, L=1.0).")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$\omega$")
plt.grid()
plt.legend()
plt.tight_layout()
plt.xlim([-8, 8])
plt.show()
```

```python
eq_points = [np.pi * x for x in range(-4, 5)]
eq_y = [0] * len(eq_points)

plt.figure(figsize=(9, 6))
plt.streamplot(xvalues, yvalues, x_dot, y_dot)
plt.scatter(eq_points, eq_y, color="red")
plt.title("Phase Space for the Pendulum (g=1.0, L=1.0)")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$\omega$")
plt.grid()
plt.xlim(-8, 8)
plt.tight_layout()
plt.show()
```
