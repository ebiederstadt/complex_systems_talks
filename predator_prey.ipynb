```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def model(Y, t, a, b, c, d):
    x = Y[0]
    y = Y[1]

    return [a * x - b * x * y, -c * y + d * x * y]
```

```python
a = 1.0
b = 1.0
c = 1.0
d = 1.0

x_values, y_values = np.meshgrid(np.arange(0, 3, 0.1), np.arange(0, 3, 0.1))

xdot = x_values - x_values * y_values
ydot = -y_values + x_values * y_values

plt.figure(figsize=(8, 6))
plt.streamplot(x_values, y_values, xdot, ydot)

plt.axvline(x=0, color="indigo")
plt.axhline(y =a / b, color="indigo")

plt.axhline(y=0, color="red")
plt.axvline(x=c/d, color="red")

plt.title("Lotka-Volterra Phase Space")
plt.tight_layout()
plt.xlabel(r"$\dot{x}$")
plt.ylabel(r"$\dot{y}$")
plt.show()
```

```python
y0 = [0.1, 1.0]
t = np.linspace(0, 50, 1000)

solution = integrate.odeint(model, y0, t, args=(a, b, c, d))

plt.figure(figsize=(8, 6))
plt.plot(solution[:, 0], label="Prey population")
plt.plot(solution[:, 1], label="Predator population")
plt.legend()
plt.title(r"Numerical Solution for Lotka-Volterra. $x_0=0.1, y_0=1.0$")
plt.show()
```
